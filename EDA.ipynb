```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
    return pd.read_csv(file_path)


df = load_dataset("/content/drive/MyDrive/Student_performance_data _.csv")
df.head(5)
```

```python
# Categorical data
# countplot
sns.countplot(data=df.head(), x='Tutoring')
df.head()['Gender'].value_counts()
```

```python
# pie chart

df.head()['Gender'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df.head()['Age'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['Age'])
```

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['Age'])
```

```python
df.describe()
```

```python
# Scatter plot(Numerical - Numerical) for the relation between two column
sns.scatterplot(x='StudyTimeWeekly', y='Absences',hue='Gender', style='Extracurricular',size='ParentalSupport' ,data=df.head())
plt.show()
```

```python
# Bar plot (Numerical - Categorical)
  # average age in pclass
sns.barplot(x='GradeClass',y='Age',hue='Gender',data=df)
```

```python
#Box plot (Numerical - Categorical)

sns.boxplot(x='Gender',y='Age',hue='GradeClass',data=df)
```

```python
# Displot

sns.distplot(df[df['Gender']==0]['Age'],hist=False)
sns.distplot(df[df['Gender']==1]['Age'],hist=False)
```

```python
# HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(df.head()['Gender'],df.head()['StudyTimeWeekly']))
```

```python
df.groupby('Gender').mean()['Tutoring']*100
```

```python
# pairplot
sns.pairplot(df,hue='Gender')
```

```python
sns.lineplot(x='Gender',y='GradeClass',data=df)
```

```python
sns.heatmap(df.pivot_table(values='Gender',index='ParentalSupport',columns='GradeClass'))
```
